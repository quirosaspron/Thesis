# Growth Functions #
## Introduction ##
Simply put, a growth function describes the relationship between some variable â€” such as time or population â€” and the growth of a quantity. For example, it can model time versus return on investment, time versus population, or the growth of healthy versus cancerous cells.

Growth functions can often be modeled through differential equations, and there are many types of them. Because different systems behave and evolve in distinct ways, we need a variety of equations to describe these processes accurately.

In what follows, Iâ€™ll explain some of the most commonly used growth models â€” how they are defined, when they are applied, and what their characteristic shapes look like.

Growth models are particularly interesting because they provide powerful tools to describe how something increases over time â€” whether itâ€™s populations, technologies, investments, companies, or even diseases.

# Exponential growth 
### Growth rate is proportional to the current size
This model describes growth where the rate of increase is directly proportional to the current amount. It assumes **unlimited resources**, leading to a J-shaped curve that increases at an ever-accelerating rate.

It is given by:

$$
\mathbf{\frac{dY}{dt} = rY}
$$

Where:
| Variable | Description |
| :--- | :--- |
| $\mathbf{r}$ | The **growth rate** controlls how fast Y grows or decays. |
| $\mathbf{Y}$ | The **unit** we are modeling. |

**Example**: Idealized bacterial growth in unlimited resources, ie the Malthusian growth model. 

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

def exponential_growth(a, r, t):
    """
    Calculates exponential growth: y = a * (1 + r)^t
    """
    return a * (1 + r)**t

def interactive_plot(growth_rate):
    """
    Plots the exponential growth curve and its derivative side-by-side.
    """
    time_periods = np.arange(0, 11) 
    initial_amount = 100
    
    final_amounts = exponential_growth(initial_amount, growth_rate, time_periods)
    
    derivative_values = growth_rate * final_amounts
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(time_periods, final_amounts, linestyle='-', color='blue')
    ax1.set_title(f'Crecimiento: $y = {initial_amount}(1 + {growth_rate:.2f})^t$')
    ax1.set_xlabel('Tiempo (t)')
    ax1.set_ylabel('Monto Final (y)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    
    ax2.plot(final_amounts, derivative_values, linestyle='-', color='red', linewidth=2)
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = rY$')
    ax2.set_xlabel('Final ammount (y)')
    ax2.set_ylabel(r'Rate of change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

r_slider = FloatSlider(min=0.01, max=0.50, step=0.01, value=0.15, description='GR (r):', continuous_update=False)

interact(interactive_plot, growth_rate=r_slider);
```

# Monomolecular Equation 
### Growth toward a fixed asymptote at a rate proportional to the distance between the curve and the asymptote (the gap)
The name of this function comes because it was first used to describe the progress of a **simple irreversible first-order chemical reaction**. 

This equation models growth that is **initially fast but continuously slows down** as it approaches a defined maximum limit or asymptote. The growth rate is proportional to the difference between the current value and the maximum limit.

The equation for the Monomolecular growth model (describing growth towards a limit) is:


$$\mathbf{\frac{dY}{dt} = r(K - Y)}$$

Where:

| Variable | Description |
| :--- | :--- |
| $\mathbf{K}$ | The **Carrying capacity** or **Limiting Value** (the asymptote). |
| $\mathbf{r}$ | The **Rate Constant** (determining how quickly $Y$ approaches $K$). |

**Example**: certain chemical reactions approaching equilibrium, or simple plant growth under limiting factors.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

def monomolecular_growth(K, r, t):
    """
    Calculates Monomolecular growth: y = K * (1 - e^(-r*t))
    K = Carrying Capacity
    r = Rate of growth
    t = time period
    """
    return K * (1 - np.exp(-r * t))

def interactive_monomolecular_plot(Growth_Rate_r):
    """
    Plots the monomolecular curve based on interactive parameters.
    """
    time_periods = np.arange(0, 10, 0.1)
    Carrying_Capacity_K = 1000
    
    
    final_amounts = monomolecular_growth(Carrying_Capacity_K, Growth_Rate_r, time_periods)
    
    derivative_values = Growth_Rate_r * (Carrying_Capacity_K - final_amounts)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(time_periods, final_amounts, linestyle='-', color='purple', label='Monomolecular Curve')
    
    ax1.axhline(y=Carrying_Capacity_K, color='r', linestyle='--', alpha=0.6, label='Carrying Capacity (K)')
    
    ax1.set_title(f'Monomolecular Growth: K={Carrying_Capacity_K:.0f}, r={Growth_Rate_r:.2f}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Amount (y)')
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.6)
    
    ax2.plot(final_amounts, derivative_values, linestyle='-', color='red')
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = r(K-Y)$')
    ax2.set_xlabel('Amount (y)')
    ax2.set_ylabel(r'Derivative ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()


r_slider = FloatSlider(min=0.1, max=1.0, step=0.05, value=0.3, description='GR (r):', continuous_update=False)

interact(interactive_monomolecular_plot, Growth_Rate_r=r_slider);
```

# A note about the exponential and monomolecular models
## Convexity vs. Concavity: Second Derivative Test

A key difference between the exponential and monomolecular growht is that exponential growth is convex (accelerating) and monomolecular growth is concave (decelerating).

### 1. Exponential Growth (Convex)

**The Equation:**
$$\frac{dY}{dt} = rY$$

**The Second Derivative (Curvature):**
To find the curvature, we differentiate the rate equation with respect to time using the chain rule:

$$\frac{d^2Y}{dt^2} = \frac{d}{dt}(rY)$$

$$= r \frac{dY}{dt}$$

Substitute $\frac{dY}{dt}$ back into the equation:

$$= r (rY)$$

$$\mathbf{\frac{d^2Y}{dt^2} = r^2 Y}$$

**Conclusion:**
Since $r^2$ is always positive and population $Y$ is positive:
$$\frac{d^2Y}{dt^2} > 0 \implies \textbf{Convex (Curves Up / Accelerating)}$$

---

### 2. Monomolecular Growth (Concave)

**The Equation:**
$$\frac{dY}{dt} = r(K - Y)$$

**The Second Derivative (Curvature):**
We differentiate the rate equation with respect to time. $K$ is a constant, so its derivative is 0.

$$\frac{d^2Y}{dt^2} = \frac{d}{dt} [r(K - Y)]$$

$$= r \left( \frac{dK}{dt} - \frac{dY}{dt} \right)$$

$$= r \left( 0 - \frac{dY}{dt} \right)$$

$$= -r \frac{dY}{dt}$$

Substitute $\frac{dY}{dt}$ back into the equation:

$$= -r [r(K - Y)]$$

$$\mathbf{\frac{d^2Y}{dt^2} = -r^2 (K - Y)}$$

**Conclusion:**
Assuming we are in the growth phase where $Y < K$, the term $(K-Y)$ is positive. However, because of the negative sign in front:
$$\frac{d^2Y}{dt^2} < 0 \implies \textbf{Concave (Curves Down / Decelerating)}$$

# Logistic Equation 
### Growth is slow early on, then it takes off in an exponential way and slows down again as you apporach the capacity
The logistic model describes sigmoid (S-shaped) growth where the rate is slow initially, increases to a maximum at the halfway point, and then slows down as it approaches a carrying capacity (K). It is the standard model for population growth limited by environmental resources. The central idea is that the environment restricts growth as the population size gets larger.

The differential equation form of the logistic equation is given by:

$$
\mathbf{\frac{dY}{dt} = r \, Y \left(1 - \frac{Y}{K}\right)}
$$

Where:

| Variable | Description |
| :--- | :--- |
| $\mathbf{r}$ | The **Growth rate**. |
| $\mathbf{K}$ | The **Carrying capacity** or **Maximum Limit**. | 


**Example**: population growth in limited habitat, spread of species.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def logistic_solution(y0, r, K, t):
    """
    Calculates the exact logistic growth solution y(t).
    """
    return K / (1 + ((K - y0)/y0) * np.exp(-r*t))

def interactive_plot(r, K):
    """
    Plots the logistic growth curve using the analytic solution.
    """
    y0 = 1 
    t = np.linspace(0, 20, 200)  
    y = logistic_solution(y0, r, K, t)
    

    dydt = r * y * (1 - y / K)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    ax1.plot(t, y, color='blue', lw=2)
    ax1.axhline(y=K, color='r', linestyle='--', label='Carrying Capacity (K)')
    ax1.set_title(f'Logistic Growth Curve\nr = {r}, K = {K}, y0 = {y0}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Population (y)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = rY(1 - \frac{Y}{K})$')
    ax2.set_xlabel('Population (y)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

r_slider = FloatSlider(min=0.01, max=1.0, step=0.01, value=0.5, description='r:')
K_slider = FloatSlider(min=5, max=100, step=1, value=20, description='K:')

interact(interactive_plot, r=r_slider, K=K_slider);
```

# Gompertz Equation 
### Growth slows more gradually than logistic, asymmetric S-shape
This model describes sigmoid (S-shaped) growth toward a maximum limit, but it introduces an asymmetry not present in the logistic model. The growth rate decreases exponentially over time, so the decline phase begins earlier and is more gradual.

It is widely used to model tumor growth and in mortality or survival analysis. It originated in actuarial science, where Benjamin Gompertz (1825) developed it to describe human mortality rates, assuming that mortality increases exponentially with age. 

The Gompertz model can be viewed as a special case of the logistic equation, describing situations where the system approaches its limit more quickly at first.
$$
\mathbf{\frac{dY}{dt} = rY \ln\left(\frac{K}{Y}\right)}
$$

Where:
| Variable | Description |
| :--- | :--- |
| $\mathbf{K}$ | The **Carrying capacity** or **Limiting Value** (the upper asymptote). |
| $\mathbf{r}$ | The **Rate Constant** (determining how quickly $y$ approaches $M$). |

**Example**: tumor growth, human mortality, microbial growth that slows over time.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def gompertz_solution(K, r, t, y0):
    """
    Calculates Gompertz growth y(t) with initial value y0.
    """
    # Calculate t0 from initial condition
    t0 = -np.log(-np.log(y0 / K)) / r
    return K * np.exp(-np.exp(-r * (t - t0)))

def interactive_plot(K, r):
    """
    Plots the Gompertz growth curve using the analytic solution.
    """
    y0 = 1
    t = np.linspace(0, 20, 200)
    y = gompertz_solution(K, r, t, y0)
    
    dydt = r * y * np.log(K / y)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, y, color='green', lw=2)
    ax1.axhline(y=K, color='r', linestyle='--', label='Carrying Capacity (K)')
    ax1.set_title(f'Gompertz Growth Curve\nr = {r}, K = {K}, y0 = {y0}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Population (y)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = rY \ln(\frac{K}{Y})$')
    ax2.set_xlabel('Population (y)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

r_slider = FloatSlider(min=0.01, max=0.5, step=0.01, value=0.15, description='r:')
K_slider = FloatSlider(min=2, max=100, step=1, value=20, description='K:')

interact(interactive_plot, K=K_slider, r=r_slider);
```

# Under the hood of the Gompertz Model

The Gompertz equation is commonly written as $\frac{dY}{dt} = r Y \ln\left(\frac{K}{Y}\right)$, by expanding the logarithm we can get some insight into the internal mechanics of the model and see how its behavior is fundamentally different from the Logistic model.


Using the identity $\ln(A/B) = \ln A - \ln B$, we can rewrite the differential equation:

$$
\frac{dY}{dt} = r Y (\ln K - \ln Y)
$$

Distributing $Y$ allows us to see two distinct terms that are competing:

$$
\frac{dY}{dt} = \underbrace{(r \ln K) \cdot Y}_{\text{Linear Impulse}} - \underbrace{r (Y \ln Y)}_{\text{ Brake}}
$$

This reveals that Gompertz growth is the net result of two opposing mathematical forces:

The term $(r \ln K) Y$ is a standard linear growth term. If this were the only term, we would see simple unbounded exponential growth. But the term $-r (Y \ln Y)$ acts as the drag or friction. The function $x \ln x$ appears frequently in thermodynamics (Gibbs entropy) and information theory. As $Y$ increases, the term $Y \ln Y$ grows **super-linearly** (it grows faster than simple $Y$). This means the "drag" eventually neutralizes the linear growth impulse exactly when $Y = K$.


The key feature of the Gompertz model is its asymmetry. The derivative of the logistic model, is a perfect parabola with a peak at $Y = 0.5K$.

We can find the exact moment of maximum growth speed by maximizing the rate function $f(Y) = Y \ln(K/Y)$.

Setting the derivative to zero:

$$
\frac{d}{dY} \left[ Y (\ln K - \ln Y) \right] = 0
$$

$$
1 \cdot (\ln K - \ln Y) + Y \cdot \left(-\frac{1}{Y}\right) = 0
$$

$$
\ln\left(\frac{K}{Y}\right) - 1 = 0 \implies \ln\left(\frac{K}{Y}\right) = 1
$$

$$
\frac{K}{Y} = e \implies \mathbf{Y = \frac{K}{e} \approx 0.368 K}
$$

### Summary Comparison
So while the Logistic Model has a quadratic breaking mehcanism ($-Y^2$), the Gompertz has a Logarithmic breaking mechanism $-Y \ln Y$. Due to this we get a differetn point of maximum growth, in the Logistic it is achived at **50%** of $K$, but in Gompertz it's at  **36.8%** ($1/e$) of $K$. The qualitative difference is that there is an asymetry in the Gompertz shown as an early peak, where we reach the a maximum point early and then we spend a long time slowly decelerating. Whereas in the Logistic it's more symetric, we slow down at the same rate as we  sped up. 



# Chanter Equation
### A hybrid of logistic & Gompertz

It extends the Gompertz model by introducing additional parameters to better fit certain real-world datasets. 

It is particularly useful in scenarios where the standard Gompertz model may not adequately describe the observed growth patterns. It adds a time-dependent exponential modifier to gap-driven growth (logistic style) making it more flexible.

It's differential form is given by:
$$
\mathbf{\frac{dY}{dt} = \mu \, Y \left(1 - \frac{Y}{B}\right) e^{-D t}}
$$


Where:
| Variable        | Description                                                                      |
| :-------------- | :------------------------------------------------------------------------------- |
| $\mathbf{B}$    | **Maximum or limiting value** (the asymptote).                                   |
| $\mathbf{\mu}$  | **Growth rate constant** (controls how fast $y$ grows initially).                |
| $\mathbf{D}$    | **Decay parameter** (controls the exponential decline of growth rate over time). |


Adding complexity to the Chanter model gives us more flexibility to fit real-world growth patterns that the simpler models canâ€™t capture. With its extra parameters, we can adjust the curve to account for slow early growth, rapid mid-growth, and a gradual approach to the maximum value, which often happens in biological or industrial processes. P Wtih this model thereâ€™s an important trade-off: more parameters make the model harder to fit and interpret, and without sufficient or complex enough data, it could very easily overfit our dataset. 

**Example**: fruit growth of cocoa where pure logistic or Gompertz didnâ€™t fit well.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def chanter_solution(y0, B, mu, D, t):
    """
    Calculates Chanter growth y(t) using the analytical solution.
    """
    return (y0 * B) / (y0 + (B - y0) * np.exp(-(mu/D) * (1 - np.exp(-D * t))))

def interactive_chanter(y0, B, mu, D):
    t = np.linspace(0, 20, 200)
    y = chanter_solution(y0, B, mu, D, t)
    
    dydt = mu * y * (1 - y/B) * np.exp(-D * t)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, y, color='blue', lw=2)
    ax1.axhline(y=B, color='r', linestyle='--', label=f'B = {B}')
    ax1.set_title(f'Chanter Growth Curve\ny0={y0}, B={B}, Î¼={mu}, D={D}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Population (y)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = \mu Y(1 - \frac{Y}{B})e^{-Dt}$')
    ax2.set_xlabel('Population (y)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

y0_slider = FloatSlider(min=0.1, max=10, step=0.1, value=1, description='y0:')
B_slider = FloatSlider(min=5, max=100, step=1, value=20, description='B:')
mu_slider = FloatSlider(min=0.01, max=1, step=0.01, value=0.2, description='Î¼:')
D_slider = FloatSlider(min=0.01, max=1, step=0.01, value=0.1, description='D:')

interact(interactive_chanter, y0=y0_slider, B=B_slider, mu=mu_slider, D=D_slider);
```

# Exponential Quadratic Equation 
### The growth rate itself changes linearly in time.
This function involves a combination of exponential growth with a quadratic component in the exponent. This structure allows the growth rate to accelerate or decelerate based on the quadratic term, providing greater flexibility than simple exponential growth. It is often used as an empirical model to fit complex, non-standard biological or ecological data.

An exponential quadratic growth model has a growth rate that changes linearly over time, meaning the rate itself accelerates or decelerates rather than staying constant:


$$
\mathbf{\frac{dY}{dt} = (\alpha + \beta t) Y}
$$



Where:
| Variable | Description |
| :--- | :--- |
| $\mathbf{\beta}$ | **Quadratic coefficient** ; controls acceleration or deceleration of growth over time. |
| $\mathbf{\alpha}$ | **Linear coefficient** ; shifts the growth curve along time. |

Growth can super-accelerate or decelerate depending on sign of $\beta$

**Example**: some technological growth where acceleration happens.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider, Checkbox

def exp_quad_solution(a, b, C, tmax, points):
    t = np.linspace(0, tmax, points)
    exponent = 0.5 * a * t**2 + b * t
    exponent = np.clip(exponent, -700, 700) 
    y = C * np.exp(exponent)
    return t, y

def update_plot(a, b, C, tmax, points, logscale):
    
    t, y = exp_quad_solution(a, b, C, tmax, points)
    
 
    dydt = y * (a * t + b)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, y, color='blue', lw=2)
    ax1.set_title(f'Solution: $y(t) = C e^{{\\frac{{a}}{{2}}t^2 + bt}}$\n$a={a:.2f}, b={b:.2f}, C={C:.2f}$')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('y(t)')
    if logscale:
        ax1.set_yscale('log')
    ax1.grid(True, linestyle='--', alpha=0.6)
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $y$: $\frac{dY}{dt} = (b + at)Y$')
    ax2.set_xlabel('y(t)')
    ax2.set_ylabel(r'Rate of change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

interact(update_plot, 
         a=FloatSlider(value=0.1, min=-1, max=1, step=0.01, description="a (quad)"),
         b=FloatSlider(value=0.2, min=-2, max=2, step=0.01, description="b (lin)"),
         C=FloatSlider(value=1.0, min=0, max=10, step=0.1, description="C (scale)"),
         tmax=IntSlider(value=10, min=1, max=200, step=1, description="t max"),
         points=IntSlider(value=400, min=50, max=2000, step=50, description="points"),
         logscale=Checkbox(value=False, description="log y-axis")
        );
```

# Von Bertalanffy equation
### net growth is anabolic minus catabolic processes, scaling with size
Primarily used in fisheries and animal growth modeling, this model is based on physiological principles related to anabolism and catabolism (building up vs. breaking down tissue). An organism's growth rate is the result of the balance between its energy-building and energy-breaking-down processes, and this balance changes predictably as the organism gets larger. 

It is often written as:

$$
\mathbf{\frac{dY}{dt} = p Y^a - q Y^b}
$$

| Parameters | Description |
| :--- | :--- |
| $\mathbf{p}$| Anabolic (growth) rate coefficient â€” represents the rate of tissue building or energy assimilation |
| $\mathbf{q}$ | Catabolic (decay) rate coefficient â€” represents the rate of tissue breakdown or energy expenditure.|
| $\mathbf{a}$ | Exponent describing how anabolism scales with organism size |
| $\mathbf{b}$ | Exponent describing how catabolism scales with organism size |

## The geometry of biological growth
$$
\frac{dY}{dt} = \underbrace{p Y^a}_{\text{Building up}} - \underbrace{q Y^b}_{\text{Breaking down }}
$$
The term that is **"Building up"** represents the production of new tissue. And the term that is **"Breaking down"** represents the cost of staying alive.

$p Y^a$ is the rate at which the organism assimilates nutrients and oxygen to create new cells, proteins, and structures. And $q Y^b$ is the rate at which the organism burns existing tissue or energy just to maintain itself (cellular respiration, heat loss, repairing damage).

The intresting thing about this is that $Y^a$ is limited by surface area. Nutrients and oxygen must pass through surfaces (intestines, gills, lungs, cell membranes), so it scales with surface area. When we have growth in a geometric object where the shape is constant, the surface of the object grows slower than the volume, to the power of $\approx 2/3$. So $a \approx 2/3$
##### Geometric Proof (Cube)

Let $L$ be the side length of a cube, $V$ the volume and $S$ the surface. 
$$V = L^3 \implies L = V^{1/3}$$
$$S = 6L^2$$
Substitute $L$ into the $S$ equation:
$$S = 6(V^{1/3})^2$$
**Result:**
$$S = 6 V^{2/3}$$
**Conclusion:** Surface Area is proportional to Volume raised to the power of **2/3**.



But $Y^b$ scales with Volume/Mass. Every single cell in the body needs energy to stay alive. Therefore, the "maintenance" cost scales with the total volume/weight (We treat volume as weight because most organism are can be thought of as basically "bags of water" the tissue density is about 1g/cm^3 and the density of the water is also 1g/cm^3. Basically the same. so weight is density times volume, if density is roughly 1g/cm^3 then we have weight (g)= 1g/cm^3 * volume(cm^3), so we get weight = volume.). Because cells occupy the entire volume of the body, this expenditure scales linealy with our total mass. So $b \approx 1$. 

So our model is telling us that whena organisms are small they have a lot of surface relative to their weight. This means they have a lot of income (production of tissue) and little expenditure (cost of energy). This makes them grow fast early on. But as they grow, the geometric reality pill starts to catch up. Our volume grows cubically while the surface grows cuadratically. Inevitably we will get to a point were our income = expenditure. That is when we will reach our maximum size. 

So according to this geometric reassoning, theory says: $a$ should be about $2/3$ ($0.67$). Although this depends on the assumption that the shape of the organism that is growing stays constant, which doesn't sound that realistic. Animals aren't perfect spheres. But actually this geometric constraint is not just theoretical. A recent study detailed in Quanta Magazine regarding shark growth confirms this model (see references at the end). It found that despite biological complexity, massive species scale strictly adhering to the "square-cube law" predicted by this equation.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact, FloatSlider

def general_vb_deriv(Y, t, p, a, q, b):
    """
    Defines the differential equation: dY/dt = p*Y^a - q*Y^b
    """
    Y_val = Y[0]
    if Y_val < 0: Y_val = 0
    
    dYdt = p * (Y_val**a) - q * (Y_val**b)
    return dYdt

def interactive_general_vb(p, a, q, b, y0):
    t = np.linspace(0, 50, 400)
    
    Y = odeint(general_vb_deriv, [y0], t, args=(p, a, q, b))
    Y = Y.flatten()
    
    dydt = p * (Y**a) - q * (Y**b)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, Y, color='teal', lw=2)
    ax1.set_title(f'General Von Bertalanffy\np={p}, a={a}, q={q}, b={b}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Size (Y)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    
    ax2.plot(Y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs Size: $\frac{dY}{dt} = pY^a - qY^b$')
    ax2.set_xlabel('Size (Y)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

p_slider = FloatSlider(min=0.1, max=5.0, step=0.1, value=3.0, description='p (anabolic):')
a_slider = FloatSlider(min=0.1, max=1.5, step=0.01, value=0.67, description='a (power):')
q_slider = FloatSlider(min=0.1, max=1.0, step=0.01, value=0.5, description='q (catabolic):')
b_slider = FloatSlider(min=0.5, max=2.0, step=0.1, value=1.0, description='b (power):')
y0_slider = FloatSlider(min=0.1, max=10, step=0.1, value=0.5, description='y0:')

interact(interactive_general_vb, p=p_slider, a=a_slider, q=q_slider, b=b_slider, y0=y0_slider);
```

# Richards Equation
### Generalization of logistic and Gompertz, controlling asymmetry with v

Also known as the generalized logistic function. Its core strength is its shape parameter, which allows it to model various types of sigmoid curves and locate the inflection point (the point of maximum growth) at any position on the curve. This makes it highly flexible for curve-fitting many growth patterns.

The Richards equation can be written as:

$$
\mathbf{\frac{dY}{dt} = r Y \bigl[1- \bigl(\frac{Y}{K}\bigl)^{v}\bigl]}
$$

Where:

| Variable | Description |
| :--- | :--- |
| $\mathbf{K}$ | **Asymptotic maximum value** (carrying capacity). |
| $\mathbf{r}$ | **Growth rate constant**, controls how fast \(\mathbf{Y(t)}\) approaches \(\mathbf{K}\). |
| $\mathbf{v}$ | **shape parameter** controls inflection, symmetry. |



**Note**: tuning ðœˆ allows the inflection point to shift; logistic is v = 1, Gompertz is limit v â†’ 0.

The Richards function adds the extra parameter ($\nu$, sometimes called $m$ or $\lambda$) to the Gompertz or Logistic form, which controls the point of inflection. By allowing this inflection parameter to be determined by the data, the model can accurately fit growth curves where the maximum rate occurs at 30%, 40%, 60%, or even 70% of the final size, making it much more biologically plausible for diverse organisms.

### Calculation of the Inflection Point 

The inflection point corresponds to the moment of maximum growth rate. To find it, we maximize the function $\frac{dY}{dt}$ with respect to $Y$.

Expand the Equation:
$$\frac{dY}{dt} = r Y - \frac{r}{K^v} Y^{1+v}$$

Differentiate with respect to $Y$ and set the derivative to zero to find the critical point:
$$\frac{d}{dY}\left(\frac{dY}{dt}\right) = r - \frac{r(1+v)}{K^v} Y^v = 0$$

Solve for $Y$:
$$r = \frac{r(1+v)}{K^v} Y^v$$

Divide by $r$ and rearrange:
$$Y^v = \frac{K^v}{1+v}$$

Taking the $v$-th root of both sides:
$$\mathbf{Y = \frac{K}{(1+v)^{1/v}}}$$

---

**Verification of Special Cases:**
* **Logistic ($v=1$):**
    $$Y = \frac{K}{(1+1)^1} = \frac{K}{2}$$
    *(Correct: The Logistic curve inflects at 50% of K)*

* **Gompertz ($v \to 0$):**
    Using the limit $\lim_{v \to 0} (1+v)^{1/v} = e$:
    $$Y = \frac{K}{e} \approx 0.368 K$$
    *(Correct: The Gompertz curve inflects at ~36.8% of K)*


**Example**: growth of trees or plants where logistic or Gompertz are insufficient.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def richards_solution_v(K, r, v, y0, t):
    """
    Calculates the analytic solution for the Richards equation:
    dY/dt = r * Y * (1 - (Y/K)^v)
    
    Analytic Solution:
    Y(t) = K / (1 + Q * exp(-r * v * t))^(1/v)
    where Q is determined by the initial condition y0.
    """
    # Calculate Q based on initial value y0
    # From y0 = K / (1 + Q)^(1/v)  ->  1 + Q = (K/y0)^v  ->  Q = (K/y0)^v - 1
    Q = (K / y0)**v - 1
    
    return K / (1 + Q * np.exp(-r * v * t))**(1/v)

def interactive_richards_v(K, r, v, y0):
    t = np.linspace(0, 50, 400)
    
    y = richards_solution_v(K, r, v, y0, t)
    
    dydt = r * y * (1 - (y / K)**v)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, y, color='teal', lw=2)
    ax1.axhline(y=K, color='r', linestyle='--', label=f'K = {K}')
    ax1.set_title(f'Richards Growth Curve\nK={K}, r={r}, v={v}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Population Y(t)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $Y$: $\frac{dY}{dt} = rY[1 - (\frac{Y}{K})^v]$')
    ax2.set_xlabel('Population Y(t)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

# Sliders for interactive parameters
K_slider = FloatSlider(min=10, max=200, step=5, value=100, description='K (Max):')
r_slider = FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description='r (Rate):')
v_slider = FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0, description='v (Shape):')
y0_slider = FloatSlider(min=1, max=50, step=1, value=5, description='y0 (Init):')

# Launch interactive plot
interact(interactive_richards_v, K=K_slider, r=r_slider, v=v_slider, y0=y0_slider);
```

# Schumacher Equation
### Variant of Gompertz controlling curvature by exponent
Primarily used in forestry, this equation models the growth and yield of even-aged timber stands. Its core is a transformation that linearizes the relationship between the natural logarithm of the volume and the reciprocal of the age. This simple, reliable structure makes it a standard for predicting forest stand attributes like height and diameter over time.

Mathematically, it is expressed as:

$$
\mathbf{\frac{dY}{dt} = r Y \ln \bigl(\frac{K}{Y}\bigl)^{v}}
$$


Where:
| Variable | Description |
| :--- | :--- |
| $\mathbf{K}$ | The **Carrying capacity** or **Limiting Value** (the upper asymptote). |
| $\mathbf{r}$ | The **Rate Constant** (determining how quickly $y$ approaches $K$). |
| $\mathbf{v}$ | **shape parameter** controls inflection, symmetry. |

**Example**: animal growth curves in livestock modeling.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact, FloatSlider

def generalized_gompertz_deriv(y, t, r, K, v):
    """
    Computes dy/dt = r * y * (ln(K/y))^v
    """
    y_val = y[0]
    

    if y_val <= 0 or y_val >= K:
        return 0
    
    try:
        dydt = r * y_val * (np.log(K / y_val)**v)
        return dydt
    except:
        return 0

def interactive_gen_gompertz(K, r, v, y0):
    t = np.linspace(0, 50, 400)
    
    # Solve ODE numerically
    y_solution = odeint(generalized_gompertz_deriv, [y0], t, args=(r, K, v))
    y = y_solution.flatten()
    
    # We use a safe calculation ignoring values where log might fail (e.g. at asymptotes)
    with np.errstate(invalid='ignore'):
        dydt = r * y * (np.log(K / y)**v)
    
    # Clean up any NaNs (Not a Number) that might occur at the very end if y hits K
    dydt = np.nan_to_num(dydt)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.plot(t, y, color='brown', lw=2)
    ax1.axhline(y=K, color='r', linestyle='--', label=f'K = {K}')
    ax1.set_title(f'Generalized Gompertz Growth\nK={K}, r={r}, v={v}')
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('Population Y(t)')
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend()
    
    ax2.plot(y, dydt, color='red', lw=2)
    ax2.set_title(r'Derivative vs $Y$: $\frac{dY}{dt} = r Y \ln(\frac{K}{Y})^v$')
    ax2.set_xlabel('Population Y(t)')
    ax2.set_ylabel(r'Rate of Change ($\frac{dY}{dt}$)')
    ax2.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

    
K_slider = FloatSlider(min=10, max=200, step=5, value=100, description='K (Cap):')
r_slider = FloatSlider(min=0.01, max=1.0, step=0.01, value=0.2, description='r (Rate):')
v_slider = FloatSlider(min=0.1, max=3.0, step=0.1, value=1.0, description='v (Shape):')
y0_slider = FloatSlider(min=0.1, max=10, step=0.1, value=1.0, description='y0 (Init):')

interact(interactive_gen_gompertz, K=K_slider, r=r_slider, v=v_slider, y0=y0_slider);
```

## Comparison table ##

| Model | Equation ($\frac{dY}{dt}$) | Inflection Point (Max Speed) | Qualitative "Personality" | Common Use Cases |
| :--- | :--- | :--- | :--- | :--- |
| **Exponential** | $rY$ | **None** (Accelerates forever) | **The Explosion.** Growth depends only on current size. Unbounded. | Early bacterial growth, compound interest, unchecked chain reactions. |
| **Monomolecular** | $r(K-Y)$ | **At $t=0$** (Start) | **The Gap Closer.** Starts at max speed and slows down as it fills the available space. No "S" shape. | Chemical kinetics, Newtonian cooling, organism length (not weight). |
| **Logistic** | $rY(1 - \frac{Y}{K})$ | **At $0.5 K$** (Symmetric) | **The Symmetric Balancer.** Perfect symmetry between early acceleration and late deceleration. | Ecology, standard population dynamics, market saturation. |
| **Gompertz** | $rY \ln(\frac{K}{Y})$ | **At $\approx 0.37 K$** ($K/e$) | **The Early Bloomer.** Asymmetric. Growth peaks early and has a long, slow deceleration phase. | Tumor growth, actuarial mortality, biological tissue growth. |
| **Chanter** | $\mu Y(1 - \frac{Y}{B})e^{-Dt}$ | **Variable** (Time-dependent) | **The Depleting Hybrid.** A Logistic curve where the growth potential *itself* decays over time (due to $e^{-Dt}$). | Plant growth where resources (like sunlight/nutrients) are actively depleting. |
| **Exp. Quadratic** | $(\alpha + \beta t)Y$ | **Variable** (Depends on $\beta$) | **The Transient.** The growth rate itself changes linearly. Can grow, peak, and collapse (if $\beta < 0$). | Short-term forecasting, drug concentrations, epidemics with changing transmission rates. |
| **Von Bertalanffy** | $pY^a - qY^b$ | **Variable** (Depends on $a,b$) | **The Metabolic Net.** Growth is explicitly defined as "Building Up" minus "Breaking Down." | Fisheries, animal weight/length, metabolic scaling laws. |
| **Richards** | $rY[1-(\frac{Y}{K})^\nu]$ | **Flexible** (at $(1+\nu)^{-1/\nu}K$) | **The Generalizer.** A mathematical chameleon that can become Logistic, Monomolecular, or Gompertz by tuning $\nu$. | Plant science, forestry, modeling data where the specific shape is unknown. |
| **Schumacher** (Gen. Gompertz) | $r Y [\ln(\frac{K}{Y})]^v$ | **Flexible** (Logarithmic) | **The Log-Controller.** Similar to Richards, but controls the "tail" specifically using the power of the logarithm. | Forestry (timber volume), complex tissue growth with specific decay rates. |

## Afterthoughts

The equations weâ€™ve been studying here, the Gompertz, Logistic, Richards, Von Bertalanffy, Morgan, etc.  
Are essentially **sigmoid or asymptotic growth models**.  
Their essential idea is:

**Growth rate depends on the current value relative to a limit**

This is not specific to biology. It appears in technology, AI, economics, and social sciences, for example in AI:

- **GPU/compute performance over time**: e.g., Mooreâ€™s law, where performance saturates due to physical limits.  
- **Data size vs. model performance**: accuracy often follows a logistic or power-law curve as you increase training data.  
- **Parameter growth in neural networks**: number of parameters vs. performance often shows diminishing returns, which is sigmoid-like.  
- **Benchmarks and efficiency**: energy consumption, inference time, or throughput can saturate asymptotically.  

In finance, growth models like Gompertz or logistic curves can describe processes with diminishing returns or saturation, such as:  

- Portfolio growth under risk constraints  
- Adoption of new financial products  
- Revenue and market penetration of companies  

They capture how gains accelerate initially but slow as limits like capital, market size, or risk are approached.

In electronic warfare (EW), many processes exhibit asymptotic or saturating behavior that growth models can capture:

- **Signal jamming effectiveness vs. power or resources**: initially, increasing jamming power rapidly degrades enemy signals, but additional power has diminishing returns, which is sigmoid-like.  
- **Sensor detection probability vs. observation time**: the chance of detecting a target often grows quickly at first and then approaches a maximum limit.  
- **Network or spectrum saturation**: interference and countermeasure effectiveness may plateau as the environment reaches maximum congestion.

In biotech, these growth models are directly relevant because many biological and technological processes show asymptotic, sigmoid, or saturating behavior:

- **Cell or microbial culture growth**: logistic, Gompertz, or Richards models describe how populations expand and plateau due to nutrient limits.  
- **Protein expression or enzyme activity**: increasing substrate or inducer concentration often shows diminishing returns, fitting a sigmoidal curve.  
- **Drug response / doseâ€“response curves**: the effect of a drug on cells or organisms often rises quickly at low doses and saturates at high doses.  
- **Bioprocess optimization**: scaling up bioreactors, optimizing yield, or predicting saturation points for resource use can be modeled using these equations.

Fundamentally, these growth models are **white-box tools**: they make explicit assumptions about how a system evolves, with parameters that have clear interpretations (e.g., maximum value, growth rate, shape). This allows us to understand and reason about the system, not just make predictions based on past data blindly like a black-box model would.  

Using these models provides a **white-box framework** to predict system behavior, allocate resources efficiently, and identify points of diminishing returns, which is critical in planning and optimization.
