## Brusselator Model Analysis (8.3.1)

The Brusselator is a simple, dimensionless model of a hypothetical chemical oscillator. The system kinetics are given by:
$$\dot{x}=1-(b+1)x+ax^{2}y$$
$$\dot{y}=bx-ax^{2}y$$
where $a, b>0$ are parameters and $x, y \ge 0$ are dimensionless concentrations.

---

### a) Fixed Points and Classification

#### 1. Finding Fixed Points ($\dot{x}=0$ and $\dot{y}=0$)

From the $\dot{y}=0$ nullcline:
$$bx - ax^2y = 0 \implies x(b - axy) = 0$$
Assuming $x \ne 0$:
$$b - axy = 0 \implies ax^2y = bx$$
Substitute $ax^2y = bx$ into the $\dot{x}=0$ equation:
$$\dot{x} = 1 - (b+1)x + ax^2y = 0$$
$$\dot{x} = 1 - (b+1)x + bx = 0$$
$$\dot{x} = 1 - x = 0 \implies x = 1$$

Substitute $x=1$ back into the condition $b - axy = 0$:
 $$b - a(1)y = 0 \implies y = \frac{b}{a}$$

 The unique fixed point is at $(x^*, y^*) = (1, \frac{b}{a})$

#### 2. Jacobian Matrix $J(x, y)$

 The Jacobian matrix elements are the partial derivatives of the system equations:
 $$J = \begin{pmatrix} \frac{\partial \dot{x}}{\partial x} & \frac{\partial \dot{x}}{\partial y} \\ \frac{\partial \dot{y}}{\partial x} & \frac{\partial \dot{y}}{\partial y} \end{pmatrix} = \begin{pmatrix} -(b+1)+2axy & ax^2 \\ b-2axy & -ax^2 \end{pmatrix}$$

#### 3. Classification at the Fixed Point $(1, \frac{b}{a})$

Evaluate $J$ at the fixed point $(1, \frac{b}{a})$. Since $ax^2y = b$ at this point, $2axy = 2b$ and $ax^2 = a$:
 $$A = J(1, \frac{b}{a}) = \begin{pmatrix} -(b+1)+2b & a \\ b-2b & -a \end{pmatrix} = \begin{pmatrix} b-1 & a \\ -b & -a \end{pmatrix}$$

* **Determinant ($\Delta$):**
    $$\Delta = \det(A) = (b-1)(-a) - (a)(-b) = -ab + a + ab = a$$
     Since $a > 0$, $\Delta > 0$

* **Trace ($\tau$):**
     $$\tau = \text{Tr}(A) = (b-1) + (-a) = b - 1 - a$$ 

#### 4. Fixed Point Classification

The stability and type of the fixed point depend on $\tau$ and $\Delta$:

*  Since $\Delta = a > 0$, the fixed point is never a saddle point.
* The classification depends on the sign of the trace $\tau$:
    *  **Stable Spiral/Node:** If $\tau < 0$ (i.e., $b < 1+a$)
    *  **Unstable Spiral/Node:** If $\tau > 0$ (i.e., $b > 1+a$)
    *  **Hopf Bifurcation:** If $\tau = 0$ (i.e., $b_c = 1+a$)



---

###  b) Sketch the Nullclines and Construct a Trapping Region

#### 1. Nullclines

* **$\dot{y}=0$ Nullcline (y-nullcline):**
     $$bx-ax^{2}y=0 \implies y = \frac{b}{ax}$$
  

* **$\dot{x}=0$ Nullcline (x-nullcline):**
     $$1-(b+1)x+ax^{2}y=0 \implies y = \frac{(b+1)x - 1}{ax^2}$$ 



#### 2. Constructing a Trapping Region

A trapping region is a bounded region where the flow on the boundary always points inward, guaranteeing that any trajectory starting inside remains inside.

**Flow Directions (General):**

* **$\dot{y}$ sign:**
    * Above $y = b/(ax)$ (y-nullcline): $\dot{y} < 0$ (flow **down**).
    * Below $y = b/(ax)$ (y-nullcline): $\dot{y} > 0$ (flow **up**).

* **$\dot{x}$ sign:**
    * Above $y = \frac{(b+1)x - 1}{ax^2}$ (x-nullcline): $\dot{x} > 0$ (flow **right**).
    * Below $y = \frac{(b+1)x - 1}{ax^2}$ (x-nullcline): $\dot{x} < 0$ (flow **left**).

###  c) Hopf Bifurcation 

A Hopf bifurcation occurs when the fixed point changes stability as a parameter is varied, and a limit cycle is born.  This happens when the trace $\tau$ is zero and the determinant $\Delta$ is positive.

1.  **Condition for $\tau = 0$**:
     $$\tau = b - 1 - a = 0 \implies b_c = 1 + a$$
    The Hopf bifurcation occurs at the critical parameter value $b_c = 1+a$.

2.  **Eigenvalues at Bifurcation ($b=b_c$)**:
    At $b=b_c$, the characteristic equation $\lambda^2 - \tau \lambda + \Delta = 0$ becomes:
    $$\lambda^2 + a = 0$$
     The eigenvalues are purely imaginary:
    $$\lambda = \pm i\sqrt{a}$$

3.  **Stability Change**:
    *  For $b < b_c$, $\tau < 0 \implies$ **stable spiral** .
    *  For $b > b_c$, $\tau > 0 \implies$ **unstable spiral** 

---

###  d) Limit Cycle Existence 

 The Poincar-Bendixson theorem is used to determine the existence of a limit cycle within the bounded trapping region found in part (b).

* **Case 1: $b \le b_c$ ($b \le 1+a$)**
     The fixed point is a stable spiral .  Since all trajectories within the trapping region are attracted to the stable fixed point, **no limit cycle exists**.

* **Case 2: $b > b_c$ ($b > 1+a$)**
     The fixed point is an unstable spiral (source).  The trajectories are pushed outward from the fixed point, but they are bounded by the trapping region.  The Poincar-Bendixson theorem guarantees that a closed orbit (limit cycle) must exist within the trapping region, surrounding the unstable fixed point.

The limit cycle exists for **$b > b_c$**.  This indicates a **supercritical Hopf bifurcation**.

---

###  e) Approximate Period of the Limit Cycle

 For $b \approx b_c$, the frequency of the limit cycle is approximated by the imaginary part of the eigenvalues at the bifurcation point.

*  At $b_c = 1+a$, the eigenvalues are $\lambda = \pm i\sqrt{a}$
*  The frequency is $\omega = \text{Im}(\lambda) = \sqrt{a}$
*  The approximate period $T$ is given by $T = 2\pi/\omega$
$$T \approx \frac{2\pi}{\sqrt{a}}$$
 This approximation is exact at the birth of the limit cycle ($b=b_c$)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a = 1.0
# The Hopf bifurcation occurs at b_c = 1 + a.
b_c = 1 + a
b = 3.5  # b > b_c for a stable limit cycle 

x0 = 1.002
y0 = b / a - 0.002
initial_conditions = [x0, y0]

t_max = 100
t = np.linspace(0, t_max, 5000)

# --- Brusselator Differential Equations ---
def brusselator(Z, t, a, b):
    """
    The Brusselator system of ODEs.
    
    Args:
        Z (array): Current state [x, y].
        t (float): Current time.
        a (float): Parameter a.
        b (float): Parameter b.
        
    Returns:
        list: The derivatives [dx/dt, dy/dt].
    """
    x, y = Z
    dxdt = 1 - (b + 1) * x + a * x**2 * y
    dydt = b * x - a * x**2 * y
    return [dxdt, dydt]

solution = odeint(brusselator, initial_conditions, t, args=(a, b))
x_sol = solution[:, 0]
y_sol = solution[:, 1]

x_star = 1.0
y_star = b / a
# Since b > b_c, this fixed point is an unstable spiral.

plt.style.use('seaborn-v0_8-darkgrid')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
plt.suptitle(f'Brusselator Simulation: Limit Cycle (a={a}, b={b})', fontsize=16, weight='bold')

ax1.plot(x_sol, y_sol, color='blue', linewidth=1.5, label='Trajectory')
ax1.plot(x_sol[-500:], y_sol[-500:], color='red', linewidth=2.5, label='Limit Cycle')
ax1.plot(x_star, y_star, 'o', color='black', markersize=5, label='Fixed Point (Unstable)')
ax1.set_title('Phase Plane: Trajectory to Limit Cycle')
ax1.set_xlabel('Concentration X')
ax1.set_ylabel('Concentration Y')
ax1.legend()

ax2.plot(t, x_sol, color='teal', label='X(t)', linewidth=2)
ax2.set_title('Time Evolution of Concentration X')
ax2.set_xlabel('Time (t)')
ax2.set_ylabel('Concentration X')

ax3.plot(t, y_sol, color='purple', label='Y(t)', linewidth=2)
ax3.set_title('Time Evolution of Concentration Y')
ax3.set_xlabel('Time (t)')
ax3.set_ylabel('Concentration Y')

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```
