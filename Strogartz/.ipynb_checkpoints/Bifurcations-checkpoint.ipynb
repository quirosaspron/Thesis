## Brusselator Model Analysis (8.3.1)

The Brusselator is a simple, dimensionless model of a hypothetical chemical oscillator. The system kinetics are given by:
$$\dot{x}=1-(b+1)x+ax^{2}y$$
$$\dot{y}=bx-ax^{2}y$$
where $a, b>0$ are parameters and $x, y \ge 0$ are dimensionless concentrations.

---

### a) Fixed Points and Classification

#### 1. Finding Fixed Points ($\dot{x}=0$ and $\dot{y}=0$)

From the $\dot{y}=0$ nullcline:
$$bx - ax^2y = 0 \implies x(b - axy) = 0$$
Assuming $x \ne 0$:
$$b - axy = 0 \implies ax^2y = bx$$
Substitute $ax^2y = bx$ into the $\dot{x}=0$ equation:
$$\dot{x} = 1 - (b+1)x + ax^2y = 0$$
$$\dot{x} = 1 - (b+1)x + bx = 0$$
$$\dot{x} = 1 - x = 0 \implies x = 1$$

Substitute $x=1$ back into the condition $b - axy = 0$:
 $$b - a(1)y = 0 \implies y = \frac{b}{a}$$

 The unique fixed point is at $(x^*, y^*) = (1, \frac{b}{a})$

#### 2. Jacobian Matrix $J(x, y)$

 The Jacobian matrix elements are the partial derivatives of the system equations:
 $$J = \begin{pmatrix} \frac{\partial \dot{x}}{\partial x} & \frac{\partial \dot{x}}{\partial y} \\ \frac{\partial \dot{y}}{\partial x} & \frac{\partial \dot{y}}{\partial y} \end{pmatrix} = \begin{pmatrix} -(b+1)+2axy & ax^2 \\ b-2axy & -ax^2 \end{pmatrix}$$

#### 3. Classification at the Fixed Point $(1, \frac{b}{a})$

Evaluate $J$ at the fixed point $(1, \frac{b}{a})$. Since $ax^2y = b$ at this point, $2axy = 2b$ and $ax^2 = a$:
 $$A = J(1, \frac{b}{a}) = \begin{pmatrix} -(b+1)+2b & a \\ b-2b & -a \end{pmatrix} = \begin{pmatrix} b-1 & a \\ -b & -a \end{pmatrix}$$

* **Determinant ($\Delta$):**
    $$\Delta = \det(A) = (b-1)(-a) - (a)(-b) = -ab + a + ab = a$$
     Since $a > 0$, $\Delta > 0$

* **Trace ($\tau$):**
     $$\tau = \text{Tr}(A) = (b-1) + (-a) = b - 1 - a$$ 

#### 4. Fixed Point Classification

The stability and type of the fixed point depend on $\tau$ and $\Delta$:

*  Since $\Delta = a > 0$, the fixed point is never a saddle point.
* The classification depends on the sign of the trace $\tau$:
    *  **Stable Spiral/Node:** If $\tau < 0$ (i.e., $b < 1+a$)
    *  **Unstable Spiral/Node:** If $\tau > 0$ (i.e., $b > 1+a$)
    *  **Hopf Bifurcation:** If $\tau = 0$ (i.e., $b_c = 1+a$)



---

###  b) Sketch the Nullclines and Construct a Trapping Region

#### 1. Nullclines

* **$\dot{y}=0$ Nullcline (y-nullcline):**
     $$bx-ax^{2}y=0 \implies y = \frac{b}{ax}$$
  

* **$\dot{x}=0$ Nullcline (x-nullcline):**
     $$1-(b+1)x+ax^{2}y=0 \implies y = \frac{(b+1)x - 1}{ax^2}$$ 



#### 2. Constructing a Trapping Region

A trapping region is a bounded region where the flow on the boundary always points inward, guaranteeing that any trajectory starting inside remains inside.

**Flow Directions (General):**

* **$\dot{y}$ sign:**
    * Above $y = b/(ax)$ (y-nullcline): $\dot{y} < 0$ (flow **down**).
    * Below $y = b/(ax)$ (y-nullcline): $\dot{y} > 0$ (flow **up**).

* **$\dot{x}$ sign:**
    * Above $y = \frac{(b+1)x - 1}{ax^2}$ (x-nullcline): $\dot{x} > 0$ (flow **right**).
    * Below $y = \frac{(b+1)x - 1}{ax^2}$ (x-nullcline): $\dot{x} < 0$ (flow **left**).

###  c) Hopf Bifurcation 

A Hopf bifurcation occurs when the fixed point changes stability as a parameter is varied, and a limit cycle is born.  This happens when the trace $\tau$ is zero and the determinant $\Delta$ is positive.

1.  **Condition for $\tau = 0$**:
     $$\tau = b - 1 - a = 0 \implies b_c = 1 + a$$
    The Hopf bifurcation occurs at the critical parameter value $b_c = 1+a$.

2.  **Eigenvalues at Bifurcation ($b=b_c$)**:
    At $b=b_c$, the characteristic equation $\lambda^2 - \tau \lambda + \Delta = 0$ becomes:
    $$\lambda^2 + a = 0$$
     The eigenvalues are purely imaginary:
    $$\lambda = \pm i\sqrt{a}$$

3.  **Stability Change**:
    *  For $b < b_c$, $\tau < 0 \implies$ **stable spiral** .
    *  For $b > b_c$, $\tau > 0 \implies$ **unstable spiral** 

---

###  d) Limit Cycle Existence 

 The Poincar-Bendixson theorem is used to determine the existence of a limit cycle within the bounded trapping region found in part (b).

* **Case 1: $b \le b_c$ ($b \le 1+a$)**
     The fixed point is a stable spiral .  Since all trajectories within the trapping region are attracted to the stable fixed point, **no limit cycle exists**.

* **Case 2: $b > b_c$ ($b > 1+a$)**
     The fixed point is an unstable spiral (source).  The trajectories are pushed outward from the fixed point, but they are bounded by the trapping region.  The Poincar-Bendixson theorem guarantees that a closed orbit (limit cycle) must exist within the trapping region, surrounding the unstable fixed point.

The limit cycle exists for **$b > b_c$**.  This indicates a **supercritical Hopf bifurcation**.

---

###  e) Approximate Period of the Limit Cycle

 For $b \approx b_c$, the frequency of the limit cycle is approximated by the imaginary part of the eigenvalues at the bifurcation point.

*  At $b_c = 1+a$, the eigenvalues are $\lambda = \pm i\sqrt{a}$
*  The frequency is $\omega = \text{Im}(\lambda) = \sqrt{a}$
*  The approximate period $T$ is given by $T = 2\pi/\omega$
$$T \approx \frac{2\pi}{\sqrt{a}}$$
 This approximation is exact at the birth of the limit cycle ($b=b_c$)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a = 1.0
# The Hopf bifurcation occurs at b_c = 1 + a.
b_c = 1 + a
b = 3.5  # b > b_c for a stable limit cycle 

x0 = 1.002
y0 = b / a - 0.002
initial_conditions = [x0, y0]

t_max = 100
t = np.linspace(0, t_max, 5000)

# --- Brusselator Differential Equations ---
def brusselator(Z, t, a, b):
    """
    The Brusselator system of ODEs.
    
    Args:
        Z (array): Current state [x, y].
        t (float): Current time.
        a (float): Parameter a.
        b (float): Parameter b.
        
    Returns:
        list: The derivatives [dx/dt, dy/dt].
    """
    x, y = Z
    dxdt = 1 - (b + 1) * x + a * x**2 * y
    dydt = b * x - a * x**2 * y
    return [dxdt, dydt]

solution = odeint(brusselator, initial_conditions, t, args=(a, b))
x_sol = solution[:, 0]
y_sol = solution[:, 1]

x_star = 1.0
y_star = b / a
# Since b > b_c, this fixed point is an unstable spiral.

plt.style.use('seaborn-v0_8-darkgrid')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
plt.suptitle(f'Brusselator Simulation: Limit Cycle (a={a}, b={b})', fontsize=16, weight='bold')

ax1.plot(x_sol, y_sol, color='blue', linewidth=1.5, label='Trajectory')
ax1.plot(x_sol[-500:], y_sol[-500:], color='red', linewidth=2.5, label='Limit Cycle')
ax1.plot(x_star, y_star, 'o', color='black', markersize=5, label='Fixed Point (Unstable)')
ax1.set_title('Phase Plane: Trajectory to Limit Cycle')
ax1.set_xlabel('Concentration X')
ax1.set_ylabel('Concentration Y')
ax1.legend()

ax2.plot(t, x_sol, color='teal', label='X(t)', linewidth=2)
ax2.set_title('Time Evolution of Concentration X')
ax2.set_xlabel('Time (t)')
ax2.set_ylabel('Concentration X')

ax3.plot(t, y_sol, color='purple', label='Y(t)', linewidth=2)
ax3.set_title('Time Evolution of Concentration Y')
ax3.set_xlabel('Time (t)')
ax3.set_ylabel('Concentration Y')

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```

```python
from ipywidgets import interact, FloatSlider


def interactive_brusselator(a, b):
    # The Hopf bifurcation occurs at b_c = 1 + a
    b_c = 1 + a
    # b > b_c for a stable limit cycle 
    x_range = np.linspace(0, 6, 100)
    y_range = np.linspace(0,max(6,2* b/a), 100)
    X, Y = np.meshgrid(x_range, y_range)


    U = 1 - (b + 1) * X + a * X**2 * Y
    V = b * X - a * X**2 * Y

    speed = np.sqrt(U**2 + V**2)
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    stream = ax.streamplot(X, Y, U, V, density = 2, color = speed, cmap='winter')
    fig.colorbar(stream.lines)
    ax.set_title('Brusselator phase portrait')
    ax.set_xlabel("X concentration")
    ax.set_ylabel("Y concentration")
    ax.plot(1, b/a, 'ro')
    
a_slider = FloatSlider(min=0.01, max=6, step=0.01, value=0.15, description='a:', continuous_update=False)
b_slider = FloatSlider(min=0.01, max=6, step=0.01, value=0.15, description='b:', continuous_update=False)

interact(interactive_brusselator, a=a_slider, b=b_slider);
```

### 8.1.10 Spruce Budworm Forest Dynamics

**a) Biological Interpretation**
* **$r_s, r_E$**: Growth rates for tree size and energy reserves, respectively. 
* **$K_s, K_E$**: Carrying capacities (maximum average tree size and max energy). 
* **$1 - \frac{S}{K_s}\frac{K_E}{E}$**: Logistic growth term; growth stops if tree size $S$ and energy $E$ are close to being maxed out. 
* **$P \frac{B}{S}$**: Consumption rate; as tree size $S$ gets larger, the effect of the constant budworm population $B$ is diluted. 

**b) Nondimensionalization**
Let $\hat{s} = \frac{S}{K_s}$, $e = \frac{E}{K_E}$, and $\tau = r_s t$. 

The system simplifies to:
$$\frac{d\hat{s}}{d\tau} = \hat{s} \left( 1 - \frac{\hat{s}}{e} \right)$$ 
$$\frac{de}{d\tau} = \delta e (1 - e) - \frac{\rho}{\hat{s}}$$ 

Where:
* $\delta = \frac{r_E}{r_s}$ 
* $\rho = \frac{PB}{r_s K_E K_s}$ 

**c) Nullclines and Fixed Points**
* **$\hat{s}$-nullcline ($\dot{\hat{s}}=0$):** $\hat{s} = 0$ or $\hat{s} = e$.  **(Note: we can't have ( $\hat{s}=0$ because we have this factor in our system: $\frac{\rho}{\hat{s}}$ )**

* **$e$-nullcline ($\dot{e}=0$):** $\hat{s} = \frac{\rho}{\delta e (1 - e)}$. 

To find fixed points, substitute $\hat{s} = e$ into the $\dot{e}$ equation:
$$\delta e^2 (1 - e) = \rho$$ 

**Analysis:**
* **Small $B$ (Small $\rho$):** The line $\rho$ intersects the function $f(e) = \delta e^2(1-e)$ at two points: $F_1 \approx (1,1)$ and $F_2 \approx (0,0)$. 
* **Large $B$ (Large $\rho$):** $\rho$ is greater than the maximum of the "hump" function. No fixed points exist. 
* **Bifurcation:** At the critical value of $B$, a **Saddle-Node Bifurcation** occurs.

```python
from ipywidgets import interact, FloatSlider

rs = 0.5
re = 1
ks = 10
ke = 120
b = 0.01
p = 2

s_range = np.linspace(0.001, 15, 1000)
e_range = np.linspace(0.001, 150, 1000)
S, E = np.meshgrid(s_range, e_range)

dsdt = rs * S * (1 - (S/ks)*(ke/E))
dedt = re * E * (1 - E/ke) - p * b/S

speed = np.sqrt(dsdt**2 + dedt**2)
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
stream = ax.streamplot(S, E, dsdt, dedt, density = 2, color = speed, cmap='winter')
fig.colorbar(stream.lines)
ax.set_title('Budworms vs the forest')
ax.set_xlabel("S(t) average size of trees")
ax.set_ylabel("E(t) energy reserve")
```

```python
from ipywidgets import interact, FloatSlider


def interactive_forest(rs, re, ks, ke, b, p):
    s_range = np.linspace(0.001, 15, 1000)
    e_range = np.linspace(0.001, 150, 1000)
    S, E = np.meshgrid(s_range, e_range)

    dsdt = rs * S * (1 - (S/ks)*(ke/E))
    dedt = re * E * (1 - E/ke) - p * b/S

    speed = np.sqrt(dsdt**2 + dedt**2)
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    stream = ax.streamplot(S, E, dsdt, dedt, density = 2, color = speed, cmap='winter')
    fig.colorbar(stream.lines)
    ax.set_title('Budworms vs the forest')
    ax.set_xlabel("S(t) average size of trees")
    ax.set_ylabel("E(t) energy reserve")

rs_slider = FloatSlider(min=0.01, max=6, step=0.01, value=0.15, description='rs:', continuous_update=False)
re_slider = FloatSlider(min=0.01, max=6, step=0.01, value=0.15, description='re:', continuous_update=False)
ks_slider = FloatSlider(min=0.01, max=100, step=0.01, value=0.15, description='ks:', continuous_update=False)
ke_slider = FloatSlider(min=0.01, max=200, step=0.01, value=0.15, description='ke:', continuous_update=False)
b_slider = FloatSlider(min=0.01, max=1000, step=0.01, value=0.15, description='b:', continuous_update=False)
p_slider = FloatSlider(min=0.01, max=6, step=0.01, value=0.15, description='p:', continuous_update=False)

interact(interactive_forest, rs=rs_slider, re=re_slider, ks=ks_slider, ke=ke_slider, b=b_slider, p=p_slider);
```

### 8.1.12 Interacting Bar Magnets

**a) Finding All Fixed Points**
The system is given by:
1. $\dot{\theta}_1 = K \sin(\theta_1 - \theta_2) - \sin \theta_1$
2. $\dot{\theta}_2 = K \sin(\theta_2 - \theta_1) - \sin \theta_2$

To find fixed points, set $\dot{\theta}_1 = 0$ and $\dot{\theta}_2 = 0$:
* From (1): $K \sin(\theta_1 - \theta_2) = \sin \theta_1$
* From (2): $-K \sin(\theta_1 - \theta_2) = \sin \theta_2$

Adding these two results gives $\sin \theta_1 + \sin \theta_2 = 0$, which implies $\theta_1 = -\theta_2$ or $\theta_1 = \theta_2 + \pi$.

**Case 1: Symmetric States ($\theta_1 = \theta_2$)**
If $\theta_1 = \theta_2$, then $\sin(0) = 0$, so $\sin \theta_1 = 0$ and $\sin \theta_2 = 0$.
Fixed points: $(0, 0)$ and $(\pi, \pi)$.

**Case 2: Anti-phase States ($\theta_1 = \theta_2 + \pi$)**
If $\theta_1 = \theta_2 + \pi$, then $\sin(\theta_1 - \theta_2) = \sin(\pi) = 0$. Again, $\sin \theta_1 = 0$ and $\sin \theta_2 = 0$.
Fixed points: $(\pi, 0)$ and $(0, \pi)$.

**Case 3: Coupled Off-Axis States ($\theta_1 = -\theta_2 = \alpha$)**
Substituting into equation (1):
$K \sin(2\alpha) = \sin \alpha \implies 2K \sin \alpha \cos \alpha - \sin \alpha = 0$
Factoring: $\sin \alpha (2K \cos \alpha - 1) = 0$.
If $\sin \alpha \neq 0$, then $\cos \alpha = \frac{1}{2K}$, which exists only if $K \ge \frac{1}{2}$.
Fixed points: $(\alpha, -\alpha)$, $(-\alpha, \alpha)$, $(\pi - \alpha, \pi + \alpha)$, and $(\pi + \alpha, \pi - \alpha)$, where $\alpha = \arccos(\frac{1}{2K})$.

**Summary of Fixed Points:**
$(0, 0)$, $(\pi, \pi)$, $(\pi, 0)$, $(0, \pi)$, and for $K \ge \frac{1}{2}$: $(\pm \alpha, \mp \alpha)$ and $(\pi \mp \alpha, \pi \pm \alpha)$.

**b) Jacobian and Stability Analysis**
The Jacobian matrix $A$ is:
$$A = \begin{pmatrix} K \cos(\theta_1 - \theta_2) - \cos \theta_1 & -K \cos(\theta_1 - \theta_2) \\ -K \cos(\theta_1 - \theta_2) & K \cos(\theta_1 - \theta_2) - \cos \theta_2 \end{pmatrix}$$

**Evaluating Stability:**
1. **At $(0, 0)$**: $A = \begin{pmatrix} K-1 & -K \\ -K & K-1 \end{pmatrix}$
   * $\text{Tr} = 2K - 2$, $\text{Det} = (K-1)^2 - K^2 = 1 - 2K$.
   * For $K < \frac{1}{2}$: $\text{Det} > 0, \text{Tr} < 0 \to$ **Stable Node**.
   * For $K > \frac{1}{2}$: $\text{Det} < 0 \to$ **Saddle Point**.

2. **At $(\pi, \pi)$**: $A = \begin{pmatrix} K+1 & -K \\ -K & K+1 \end{pmatrix}$
   * $\text{Tr} = 2K + 2$, $\text{Det} = 2K + 1$.
   * $\text{Det} > 0, \text{Tr} > 0 \to$ **Unstable Node**.

3. **At $(\pi, 0)$**: $A = \begin{pmatrix} -K+1 & K \\ K & -K-1 \end{pmatrix}$
   * $\text{Tr} = -2K$, $\text{Det} = K^2 - 1 - K^2 = -1$.
   * $\text{Det} < 0 \to$ **Saddle Point**.

4. **At $(0, \pi)$**: $A = \begin{pmatrix} -K-1 & K \\ K & -K+1 \end{pmatrix}$
   * $\text{Tr} = -2K$, $\text{Det} = K^2 - 1 - K^2 = -1$.
   * $\text{Det} < 0 \to$ **Saddle Point**.

5. **At $(\alpha, -\alpha)$**: (Using $\cos \alpha = \frac{1}{2K}$ and $\cos(2\alpha) = \frac{1}{2K^2}-1$)
   $A = \begin{pmatrix} -K & K - \frac{1}{2K} \\ K - \frac{1}{2K} & -K \end{pmatrix}$
   * $\text{Tr} = -2K$, $\text{Det} = K^2 - (K - \frac{1}{2K})^2 = 1 - \frac{1}{4K^2}$.
   * Since $K > \frac{1}{2}$, $\text{Det} > 0, \text{Tr} < 0 \to$ **Stable Node**.

6. **At $(\pi - \alpha, \pi + \alpha)$**: (Using $\cos(\pi - \alpha) = -\frac{1}{2K}$)
   $A = \begin{pmatrix} K & K - \frac{1}{2K} \\ K - \frac{1}{2K} & K \end{pmatrix}$
   * $\text{Tr} = 2K$, $\text{Det} = K^2 - (K - \frac{1}{2K})^2 = 1 - \frac{1}{4K^2}$.
   * Since $K > \frac{1}{2}$, $\text{Det} > 0, \text{Tr} > 0 \to$ **Unstable Node**.

**Bifurcation Type:**
At **$K = 1/2$**, a **Supercritical Pitchfork Bifurcation** occurs.
The $(0,0)$ state loses stability and splits into two stable branches, $(\alpha, -\alpha)$ and $(-\alpha, \alpha)$. Simultaneously, an unstable bifurcation occurs involving the other two off-axis points $(\pi \mp \alpha, \pi \pm \alpha)$ emerging from a state near $(\pi, \pi)$.

**Notation: A branch refers to a curve of fixed points in the phase space that exists for a specific range of the parameter $K$.**

```python
import numpy as np
import matplotlib.pyplot as plt

k = 0.2  

t_range = np.linspace(0, 2 * np.pi, 100)

TH1, TH2 = np.meshgrid(t_range, t_range)

dTH1dt = k * np.sin(TH1 - TH2) - np.sin(TH1)
dTH2dt = k * np.sin(TH2 - TH1) - np.sin(TH2)

speed = np.sqrt(dTH1dt**2 + dTH2dt**2)

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
stream = ax.streamplot(t_range, t_range, dTH1dt, dTH2dt, 
                       density=2, color=speed, cmap='winter')

fig.colorbar(stream.lines, label='Flow Speed')
ax.set_title(f'Interacting Bar Magnets Phase Portrait (K={k})')
ax.set_xlabel(r"$\theta_1$")
ax.set_ylabel(r"$\theta_2$")
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(0, 2 * np.pi)

plt.show()
```
